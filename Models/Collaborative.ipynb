### COLLAB

```python
md = Rating.merge(Books,on="ISBN")
```

```python
x = md.groupby("User-ID").count()["Book-Title"].reset_index()
rater = x[x["Book-Title"]>=200].set_index("User-ID").index
```

```python
newdf = md[md["User-ID"].isin(rater)]
```

```python
y = newdf.groupby("Book-Title").count()["Publisher"].reset_index()
```

```python
nicebooks = y[y["Publisher"]>=50].set_index("Book-Title").index
```

```python
newdf = newdf[newdf["Book-Title"].isin(nicebooks)]
```

```python
pt = pd.pivot_table(newdf,index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
pt.shape
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
score = cosine_similarity(pt)
```

```python
def recommend(name):
        index = np.where(pt.index==name)[0][0]
        distance = score[index]
        items = sorted( list(enumerate(distance)) ,key=lambda x:x[1],reverse=True)[1:6]
        data=[]
        for x in items:
            item =[]
            a =Books[Books["Book-Title"]==pt.index[x[0]]]
            a.drop_duplicates("Book-Title",inplace=True)
            item.extend(list(a["Book-Title"].drop_duplicates()))
            item.extend(list(a["Book-Author"].drop_duplicates()))
            item.extend(list(a["Year-Of-Publication"].drop_duplicates()))
            item.extend(list(a["Publisher"].drop_duplicates()))
            item.extend(list(a["Image-URL-M"].drop_duplicates()))
            data.append(item)
        return data
```

```python
b = recommend("Message in a Bottle")
print(b)
```
